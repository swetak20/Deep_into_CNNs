```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('train.csv')
data = data.drop(columns = "id")
```

```python
train_set = data.drop('target',axis=1)
train_target = data['target']
```

```python
mean = train_set.mean(axis=0)
std = train_set.std(axis=0)
train_set = (train_set - mean)/std
```

```python
trt_df = pd.DataFrame(train_target)
trt_df = pd.get_dummies(train_target, columns = "target",dtype = int)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_set, trt_df)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
test_data = pd.read_csv("test.csv")
td_id = test_data['id']
td = test_data.drop('id',axis = 1)
mean = td.mean(axis=0)
std = td.std(axis=0)
td = (td - mean)/std
td
```

```python
td_np = td.to_numpy()
td_np
```

```python
pred = reg.predict(td)
pred
```

```python
pred_df = pd.DataFrame(pred)
pred_df
```

```python
from sklearn import preprocessing

x = pred_df.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
pred_df = pd.DataFrame(x_scaled)
pred_df
```

```python
td_id.columns = ['id']
td_id = pd.DataFrame(td_id)
td_id = td_id.to_numpy()
pred_df.insert(0,'id',td_id,True)
pred_df
```
